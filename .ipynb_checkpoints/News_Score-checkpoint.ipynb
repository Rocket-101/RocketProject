```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA
import pandas as pd
```

```python
#import data
news_df = pd.read_csv('newscleaned2.csv')
```

```python
# Scoring news with sentiment analysis using nltk
sia = SIA()
results = []

for line in news_df['News_title']:
    pol_score = sia.polarity_scores(line)
    pol_score['News_title'] = line
    results.append(pol_score)

score2 = [item['compound'] for item in results]
```

```python
#Combine Data
variables_df = pd.read_csv('df_combine.csv')
variables_df['News_score1'] = variables_df['sentiment']
variables_df['News_score2'] = score2
variables_df.head()
```

```python
variables_df = variables_df[['Attributes','High','Low','Open','Close','Volume','Adj Close','Daily Growth','LabledY','HighMinusLow','CloseMinusOpen','sentiment','News_score1','News_score2']]
```

```python
variables_df.columns = ['Date','High','Low','Open','Close','Volume','Adj Close','Daily Growth','LabledY','HighMinusLow','CloseMinusOpen','sentiment','News_score1','News_score2']
```
