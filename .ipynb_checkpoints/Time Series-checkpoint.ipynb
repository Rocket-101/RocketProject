```python
import pandas as pd
#import index data
market_df = pd.read_csv("LabeledY.csv")[['Attributes', 'High','Low','Open','Close','Volume','Adj Close','Daily Growth','LabledY']]
market_df.head()
```

```python
#Calculate Moving Average
market_df['7MA'] = market_df['Adj Close'].rolling(window=7).mean()
market_df['15MA'] = market_df['Adj Close'].rolling(window=15).mean()
```
